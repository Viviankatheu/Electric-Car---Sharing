### 1.61 ) Finding out the most popular hour for picking up a shared electric car In Paris in April, 2018

```python
hr = df1[df1['CITY']== 'Paris']
pophr = hr.groupby(['ID', 'HOUR'])[['BLUECAR COUNTER']].sum()
pophr.head()
```

```python
pophr['BLUECAR_USAGE'] = pophr['BLUECAR COUNTER'].diff()
pophr.head()
```

```python
pop = pophr[pophr['BLUECAR_USAGE'] < 0]
pop.head()
```

We will create a new column to represent the difference in cars at the station

```python
#Let's find the difference using the diff method

df1['BLUECAR_USAGE'] = df1['BLUECAR COUNTER'].diff()

#create a column for utilib usage and another for utilib 1.4 usage
df1['UTILIB_USAGE'] = df1['UTILIB COUNTER'].diff()
df1['UTILIB_1.4_USAGE'] = df1['UTILIB 1.4 COUNTER'].diff()

df1.head()
```

```python
#looking for missing values
df1.isnull().sum()
```

```python
#We shall fill the missing values with 0

df2 = df1.fillna(0)
df2.head()
```

```python
df2
```

#### 1.61 a) Most popular hour for picking up a blue car

```python
#Let's find out the most popular hour for picking up a blue car
#Note that the usage column is negative when a car has been picked

df = df2[(df2['CITY'] == 'Paris') & (df1['BLUECAR_USAGE'] < 0)]
 
bluecar = df.groupby(['HOUR'])[['BLUECAR_USAGE']].sum()
bluecar.head()
```

```python
bluecar1 = bluecar.sort_values('BLUECAR_USAGE', ascending = True)
bluecar1
```

Therefore, the most popular hour for picking up a blue car is 2100hrs. 231 cars are picked at that time.

#### 1.61 b) Most popular hour for picking up a utilib

```python
ut= df.groupby('HOUR')[['UTILIB_USAGE']].sum()
ut.head()
```

```python
#We will now sort the values in ascending order
utilib = ut.sort_values('UTILIB_USAGE', ascending = True)
utilib.head()
```

Most utilibs are picked up at 1900 and 2200 hours meaning these are the most popular hours for picking utilibs. 8 utilibs are picked up at this time

####1.61 c) Most popular hour for picking up utilib 1.4

```python
#sorting these values
uti = u.sort_values('UTILIB_1.4_USAGE', ascending = True)
uti
```

Therefore, the most popular hour for picking up a utilib 1.4 is 0000 hrs. 13 of them are picked up at this time.
